```python
import pandas as pd
import os
from IdMapping import clean_provider_ids
```

```python
# Define the path to the Excel file
folder_path = "../BoS (Business Objects) Raw Data Reports - Deidentified"
file_name = "11495 RAW Client Data Export v3_EE Workflow.xlsx"
file_path = os.path.join(folder_path, file_name)
```

```python
# Get the program from the file name
program_name = file_name.split(" ")[0]
print(f"We are looking at Program {program_name} data")
```

```python
homelessness = pd.read_excel(file_path, sheet_name="EE UDES")
# Clean EE Provider ID
homelessness = clean_provider_ids(homelessness)
homelessness.head()
```

```python
homelessness = homelessness.rename(columns={'Prior Living Situation(43)': 'Prior Living Situation','Regardless of where they stayed last night - Number of times the client has been on the streets, in ES, or SH in the past three years including today(12481)': 
                        'Times Homeless (3 years)', 'Total number of months homeless on the street, in ES or SH in the past three years(12482)': 'Months Homeless (3 years)'})
homelessness["Months Homeless (3 years)"] = homelessness["Months Homeless (3 years)"].replace({"One month (this time is the first month) (HUD)": "1", "More than 12 months (HUD)": "More than 12"})
homelessness = homelessness.fillna("Not Specified")

specified_columns = ['Prior Living Situation', 'Times Homeless (3 years)', 'Months Homeless (3 years)', 'Provider Abbrev', 'Provider Tree']
homelessness_selected = homelessness[specified_columns]

homelessness_selected.head()
```

```python
prior_living_counts = homelessness_selected.groupby(["Provider Abbrev", "Provider Tree"])['Prior Living Situation'].value_counts(dropna=True).reset_index(name="Count")
prior_living_counts
```
